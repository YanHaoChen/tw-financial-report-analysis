```python
from pymongo import MongoClient
from pymongo import DESCENDING

from urllib.parse import quote_plus
```

```python
uri = "mongodb://%s:%s@%s/?authSource=stock" % (
    quote_plus('airflower'), quote_plus('money'), 'localhost')
client = MongoClient(uri)
```

```python
results = list(client.stock.dailyTransactions.find({'stockCode': '0050'}).sort('intDate', DESCENDING))
```

```python
print(results[0], results[1])
```

```python
def total_rise(days):
    count = 0
    for day in days:
        count += 1 if day['closingDiff'] > 0 else 0
        
    return count

data = [   
    {
        # target
        'dDayPriseRise': 1 if results[i]['closingDiff'] > 0 else 0,
        'dDayClosingDiff': results[i]['closingDiff'],
        # summary
        # 'totalRise': total_rise(results[i+1:i+3]),
        # day1 columns
        'd1TradeVolume': results[i+1]['tradeVolume'],
        'd1Amount': results[i+1]['amount'],
        'd1OpeningPrice': results[i+1]['openingPrice'],
        'd1AgoHighestPrice': results[i+1]['highestPrice'],
        'd1AgoLowestPrice': results[i+1]['lowestPrice'],
        'd1AgoClosingPrice': results[i+1]['closingPrice'],
        'd1AgoClosingDiff': results[i+1]['closingDiff'],
        'd1AgoTransaction': results[i+1]['transaction'],
        'd1AgoMaxDiff': round(results[i+1]['highestPrice'] - results[i+1]['lowestPrice'], 2),
        'd1DayPriseRise': 1 if results[i+1]['closingDiff'] > 0 else 0,
    }
    for i in range(0, len(results)-3)
]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.corr()
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

scale_target = [
    'd1TradeVolume',
    'd1Amount',
    'd1OpeningPrice',
    'd1AgoHighestPrice',
    'd1AgoLowestPrice',
    'd1AgoClosingPrice',
    'd1AgoClosingDiff',
    'd1AgoTransaction',
    'd1AgoMaxDiff',
]

scaled = pd.DataFrame(scale.fit_transform(df[scale_target]), columns=scale_target)
scaled['dDayPriseRise'] = df['dDayPriseRise']
scaled
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled[scale_target], scaled['dDayPriseRise'], test_size=0.3, random_state=0)
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)

test_count = X_test.shape[0]
print("Number of labeled points out of a total %d points : %d" % (test_count, (y_test == y_pred).sum()))

from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
```

```python
print(f'     : pn,pp')
print(f'an:  {tn}, {fp}')
print(f'ap:  {fn}, {tp}')
```
